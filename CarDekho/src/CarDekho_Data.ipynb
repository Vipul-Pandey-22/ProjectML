I'm using IQR Based Filtering coz my data is skewed

```python
sns.boxplot(df['selling_price'])
```

### Finding the IQR

```python
percentile25 = car_dataframe['selling_price'].quantile(0.25)
percentile75 = car_dataframe['selling_price'].quantile(0.75)
```

```python
print(percentile25)
print(percentile75)
```

```python
# Finding the upper and lower limits
iqr = (percentile75 - percentile25)
print(iqr)
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr
```

```python
print("upper limit: ", upper_limit)
print("lower limit: ", lower_limit)
```

```python
# Finding the outliers
car_dataframe[car_dataframe['selling_price'] > upper_limit]
# df[df['selling_price'] < lower_limit]
```

```python
car_dataframe[car_dataframe['selling_price'] < lower_limit]
```

```python
new_car_df = car_dataframe[car_dataframe['selling_price'] < upper_limit]
new_car_df.shape
```

```python
sns.boxplot(new_car_df['selling_price'])
```

```python
sns.distplot(new_car_df['selling_price'])
plt.show()
```

## Now Perform Capping

```python
new_df_cap = car_dataframe.copy()
new_df_cap['selling_price'] = np.where(
    new_df_cap['selling_price'] > upper_limit,
    upper_limit,
    np.where(
        new_df_cap['selling_price'] < lower_limit,
        lower_limit,
        new_df_cap['selling_price']))
```

# Train Test Split

```python
X = new_df_cap.drop('selling_price', axis = 1)
y = new_df_cap['selling_price']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.29, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

```python
plt.subplots(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.boxplot(data=X_train)
plt.title('X_train Before Scaling')
plt.subplot(1, 2, 2)
sns.boxplot(data=X_train_scaled)
plt.title('X_train After Scaling')
```

```python
# Make Predict
y_pred = lr.predict(X_test_scaled)
y_pred
```

# Performance Matrix

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Find MAE, MSE, RMSE
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("Mean Absolute Error :", mae)
print("Mean Squared Errir :", mse)
print("Root Mean Squared Error :", rmse)
```

```python
score = r2_score(y_test,y_pred)
print("R2 Score :", score)
```

# Cross Validation

### Lasso Validation

```python
from sklearn.linear_model import LassoCV
lassoCV = LassoCV(cv=5)
lassoCV.fit(X_train_scaled, y_train)
lassoCV_pred = lassoCV.predict(X_test_scaled)
plt.scatter(y_test, lassoCV_pred)
```

```python
from sklearn.linear_model import RidgeCV
ridgeCV = RidgeCV(cv=5)
ridgeCV.fit(X_train_scaled, y_train)
ridgeCV_pred = ridgeCV.predict(X_test_scaled)
plt.scatter(y_test, ridgeCV_pred)
ridgeCV_mae = mean_absolute_error(y_test, ridgeCV_pred)
ridgeCV_mse = mean_squared_error(y_test, ridgeCV_pred)
ridgeCV_rmse = np.sqrt(mse)
print("Mean Absolute Error For RidgeCV:", ridgeCV_mae)
print("Mean Squared Error For RidgeCV:", ridgeCV_mse)
print("Root Mean Squared Error For RidgeCV:", ridgeCV_rmse)
# Check Score
ridgeCV_r2_score = r2_score(y_test, ridgeCV_pred)
print("R2 Score For RidgeCV:", ridgeCV_r2_score)
```

```python
from sklearn.linear_model import ElasticNetCV
elasticNetCV = ElasticNetCV(cv=5)
elasticNetCV.fit(X_train_scaled, y_train)
elasticNetCV_pred = elasticNetCV.predict(X_test_scaled)
plt.scatter(y_test, elasticNetCV_pred)
elasticNetCV_mae = mean_absolute_error(y_test, elasticNetCV_pred)
elasticNetCV_mse = mean_squared_error(y_test, elasticNetCV_pred)
elasticNetCV_rmse = np.sqrt(mse)
print("Mean Absolute Error For ElasticNetCV:", elasticNetCV_mae)
print("Mean Squared Error For ElasticNetCV:", elasticNetCV_mse)
print("Root Mean Squared Error For ElasticNetCV:", elasticNetCV_rmse)
# Check Score
ridgeCV_r2_score = r2_score(y_test, elasticNetCV_pred)
print("R2 Score For ElasticNet CV:", ridgeCV_r2_score)
```
