```python
# Remove Outliers By IQR Method
def remove_outliers(col):
    Q1 = np.percentile(df[col], 25)
    Q3 = np.percentile(df[col], 75)
    IQR = Q3 - Q1
    return df[(df[col] >= (Q1-1.5*IQR)) & (df[col] <= (Q3 + 1.5*IQR))]
```

```python
sns.boxplot(remove_outliers(load_california.feature_names[0])['MedInc'])
```

```python
# Remove Outliers by using Z Score
def Z_Score(data, col):
    mean = data[col].mean()
    std = data[col].std()
    lower_limit = mean - 1.9*std
    upper_limit = mean + 1.9*std
    data[col] = np.where(
        df[col] > upper_limit,
        upper_limit,
        np.where(
            df[col] < lower_limit,
            lower_limit,
            df[col]
        )
    )
    # df = data[(data[col] >= lower_value) & (data[col] <= upper_value)]
    # print(df.shape)
    return data
```
