```python
mean = dataset.MedHouseVal.mean()
mean
```

```python
std = dataset.MedHouseVal.std()
std
```

```python
lower_cut, upper_cut = mean -2 * std , mean +2 * std
print("lower cut, upper cut :", lower_cut, upper_cut)
```

```python
df_clean = dataset[(dataset['MedHouseVal'] > lower_cut) & (dataset['MedHouseVal'] < upper_cut)]
df_clean.head()
```

```python
X = dataset.drop('MedHouseVal', axis=1)
y = dataset['MedHouseVal']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
std_scaler = StandardScaler()
min_max_scaler = MinMaxScaler()
```

```python
# StandaradScaler
X_train_std_scaler = std_scaler.fit_transform(X_train)
X_test_std_scaler = std_scaler.transform(X_test)
```

```python
# MinMaxScaler
X_train_min_max_scaler = min_max_scaler.fit_transform(X_train)
X_test_min_max_scaler = min_max_scaler.transform(X_test)
```

```python
# For StaStandardScaler

plt.subplots(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.boxplot(data=X_train)
plt.title('X_train Before Scaling')
plt.subplot(1, 2, 2)
sns.boxplot(data=X_train_std_scaler)
plt.title('X_train After Scaling')
```

```python
# For MinMaxScaler

plt.subplots(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.boxplot(data=X_train)
plt.title('X_train Before Scaling')
plt.subplot(1, 2, 2)
sns.boxplot(data=X_train_min_max_scaler)
plt.title('X_train After Scaling')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
# Checking For Standard Scaler
lr.fit(X_train_std_scaler, y_train)
```

```python
y_pred = lr.predict(X_test_std_scaler)
```

```python
plt.scatter(y_test, y_pred)
# plt.plot(y_test, y_pred)
```

```python
# Calculate Error
residuals = y_test - y_pred
print(residuals)
```

```python
## Plot this residuals or error with the help of seaborn
sns.displot(residuals, kind='kde')
```

```python
## Create a Scatter plot with respect to residuals and prediction
plt.scatter(y_pred, residuals)

# Note: this data should be uniformaly distributed it means your data should not create any kind pattern 
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)
print("mse:", mse)
print("mae:" ,mae)
print("rmse:" ,rmse)
```

```python
score = r2_score(y_test, y_pred)
score
```
