## Data Cleaning

```python
df[df.isnull().any(axis=1)]
```

```python
df.loc[:122, "Regions"] = 0
df.loc[122:, "Regions"]=1
df = df
```

```python
df
```

```python
df.info()
```

```python
df['Regions'] = df['Regions'].astype(int)
```

```python
df.info()
```

### Removing the null values

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(124).reset_index(drop = True)
```

```python
df.info()
```

```python
# Removing Spaces from columns
df.columns = df.columns.str.strip()
df.columns
```

#### Converting Dtype of feature name

```python
for i in df.columns:
    if i in ['day', 'month', 'year', 'Temperature', 'RH', 'Ws']:
        df[i] = df[i].astype(int)
```

```python
df.info()
```

```python
features = [cols for cols in df.columns if df[cols].dtype == 'O'] 
features
```

```python
for object_fearure in features:
    if object_fearure != 'Classes':
        df[object_fearure] = df[object_fearure].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
# Let's save the cleaned dataset
df.to_csv('Algerian_Fire_Cleaned_Dataset.csv',index=False)
```

# Exploratory Data Analysis

```python
# Dropping day, month, and year features
df_copy = df.drop(['day', 'month', 'year'], axis = 1)
```

```python
df_copy
```

```python
# Encoding the Categorical variables in Classes
df_copy['Classes'] = df_copy['Classes'].astype(str)
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'), 0, 1)
```
